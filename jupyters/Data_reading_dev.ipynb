```python
### One hot vector
```

```python
def one_hot(vec, basis_functions):
    out = np.zeros(len(basis_functions))
    for i in vec:
        out[i] = 1
    return out
```

### Support Reading

```python
f = open(r"C:\Users\lbg\OneDrive - CSEM S.A\Bureau\Pytorch\NEW_EQ_LEARN\Data\support.txt", "r")
support = []
for i,x in enumerate(f):
    if i>0:
        support.append(float(x))
```

### Basis Mapping

```python
basis_functions = {'sin(x)':0,'exp(x)':1, 'x':2, 'exp(-x)':3, '1/x':4, 'sin(x**2)':5, 'exp(-x**2)':6,'1':7}
```

### Reading

```python
dir = r"C:\Users\lbg\OneDrive - CSEM S.A\Bureau\Pytorch\NEW_EQ_LEARN\Data\Polynomials\Pol1"
files = []
for (dirpath, dirnames, filenames) in walk(dir):
    files.extend(filenames)
    break
    
X = np.zeros([len(files),len(support)])
Y_complexity = np.zeros([len(files)])
#GT_function = np.zeros([len(files)]).astype('str')
GT_function = []
labels_one_hot = np.zeros([len(files),len(basis_functions)])


for i in range(len(files)):
    print(i)
    new_dir = os.path.join(dir,files[i])
    data = []
    with open(new_dir,"r") as f:
        data = f.readlines()
    X[i,:] = np.array(data[-998:]).astype('float')
    Y_complexity[i] = int(data[0][-2])
    x = Symbol('x')
    #GT_function[i] = lambdify(x, data[1][:-1])
    print(data[1][:-1])
    GT_function.append(data[1][:-1])
    funcs = []
    for k in range(3,3+int(data[2][-2])):
        funcs.append(data[k][:-1])

    labels = [basis_functions[funcs[j]] for j in range(len(funcs))]
    labels_one_hot[i,:] = one_hot(labels, basis_functions)
```
